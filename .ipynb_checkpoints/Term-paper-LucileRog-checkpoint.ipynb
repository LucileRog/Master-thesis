# Introduction

The latest report on food security found that the share of undernourished people has fallen down to 11\% of worldwide population in 2015 (FAO, 2015). Yet, there is a growing concern that global warming and changes in population may reverse the trend. The FAO estimates that total food production must increase by 70\% in order to feed the world in 2050(FAO, 2009).

This is preoccupying when most of arable lands are already used and a majority already suffer from erosion (Speth et al, 1994). One option may be to reduce the intake of resource-inefficient types of food such as meat-based commodities. Livestock accounted for 80\% of agricultural land use in 2006 (FAO, 2006) while it supplied only about 13\% of the calories consumed in the world (FAO, 2001). 

The relative land-inefficiency of meat-based products is mostly due to the reliance of intensive livestock farming on crops for feed purposes. Crop yields are however likely to be disrupted by climate change (Parry et al, 2004). Yet, intensive livestock farming already represented 45\% of the world's meat production in 2003 (Fao, 2006) and the share is likely to grow with meat demand.

On the other side, demand is expected to rise dramatically: by 70\% from now to 2050 (FAO, 2011). The reasons are demographic growth, increasing income and changes in Asian diets towards more meat-based products.  

As a consequence, international trade patterns and prices in food products may profoundly change in the future. Many regions will increase their reliance on trade, either as net exporters or net importers. Asia for example has been the fastest growing net importer since 2007, whereas Latin America has become the largest net exporter of food (FAO, 2015).

While future food security seems to require a diet shift towards less meat and dairy consumption, the amplitude of this shift remains unclear, as well as the instruments to achieve such a change. Will food commodities prices adjust so that food security is achieved in the future without any public intervention?

To provide some answers to these questions, I propose to model a hypothetical world in trade where meat is only produced through intensive farming, and all consumers have same preferences. The goal is to derive the conditions under which the food security constraint is satisfied.

# The model

This is a Ricardian trade model with 1 input (land), and two commodities (cereals and meat) assuming perfect competition. Countries have comparative advantages in terms of land endowments and temperatures, but have same representative agent with same preferences, and same technologies

## Autarky case

### Demand
In country $n$, demand consists in a representative consumer who has to choose his consumption of cereals $q_{c,n}$ and meat $q_{m,n}$ with a CES utility function, and the following problem:
$$ \max_{\left\lbrace q_{c,n} \geq 0, \hspace{2mm} q_{m,n} \geq 0 \right\rbrace} \left\lbrace U(q) = 
\Big[ q_{c,n}^{\frac{\sigma -1}{\sigma}} + q_{m,n}^{\frac{\sigma -1}{\sigma}} \Big]^{\frac{\sigma}{\sigma -1}} \right\rbrace
$$
Subject to:
$$ q_{c,n} + q_{m,n} p = y_n $$
The price of cereals is normalized to 1, and $p$ stands for the relative price of meat on cereals.

Because land is the only input, income in country $n$ is: $y_n = p_L \overline(L)_n$, where $\overline{L}_n$ is the total available land for agriculture in country $n$, and $p_L$ is the price of land.

The analytical solution to this maximization problem is the classical:
$$ \frac{q_{m,n}^\ast}{q_{c,n}^\ast} = p^{-\sigma} $$

### Supply
Cereal farmers have the production function:
$$ Q_{c, n} (L_{c, n}) = L_{c, n}.\theta_c(d_n) $$
Cereal production requires only land $L_{c, n}$. $d_n = | T^\ast - T_n | $ is the distance of country $n$'s temperature $T_n$ compared to the optimal temperature for growing crops $T^\ast$. $\theta_c(d_n)$ is the function of crop yields, linearly decreasing with $d_n$: $\theta_c (d_n) = \gamma_1 T^\ast - \gamma_2 d_n$.

Livestock farmers have the production function: 
$$ Q_{m, n} (L_{m, n}; K_{m, n})= \min \left\lbrace a L_{m, n} ; b K_{m, n} \right\rbrace $$
Meat production requires both land $L_{m, n}$ and cereals $K_{m, n}$ bought from cereal producers. Land and cereals are perfect complements in meat producers' function. $a$ is the amount of meat units that can be produced with one unit of land, and $b$ with one unit of cereals. Because we assume the technology used is intensive livestock farming, we assume $a\geq 1$ and $b \leq 1$. 

The perfect competition assumption yields prices:
$$ p_L = \theta_c({d_n})  \hspace{5mm} \text{and} \hspace{5mm} p^\ast = \frac{\theta_c({d_n})}{a} + \frac{1}{b}$$
Note that the relative price of meat is a constant depending on technology and temperature.

### Equilibrium
The equilibrium must satisfy the following constraints:
$$ L_{c, n} + L_{m, n} = \overline{L}_n $$
$$ Q_{c, n} (L_{c, n}) = q_{c, n} + K_{m, n} $$
$$ Q_{m, n} (L_{c, n}; K_{m, n}) = q_{m, n} $$
Note that for simplicity, we assumed the full employment of resources.

Using the fact that meat producers will equalize $a L_{m, n}$ and $b K_{m, n}$ to avoid useless costs, we can derive $Q_{m,n}$ as a function of $q_{c,n}$ and then obtain the following optimal quantities by equating relative demand to relative supply:
$$ q_{c, n}^\ast = \theta_c(d_n) \overline{L}_n \frac{1}{(p^\ast)^{1-\sigma} +1} \hspace{5mm} \text{and} \hspace{5mm} q_{m,n}^\ast = \theta_c(d_n) \overline{L}_n \frac{(p^\ast)^{-\sigma}}{(p^\ast)^{1-\sigma} +1}$$
If $p^\ast \geq 1$, $q_{c, n}^\ast$ increases and $q_{m, n}^\ast$ decreases with $\sigma$. On the contrary if $p^\ast \leq 1$, $q_{c, n}^\ast$ decreases and $q_{m, n}^\ast$ increases with $\sigma$.

Because we assumed $a\geq 1$ and $b \leq 1$, and because $p^\ast = \frac{\theta_c({d_n})}{a_m} + \frac{1}{b_m} \geq 0 $, we must have $p^\ast \geq 1$, therefore have meat consumption decrease and cereal consumption increase with CES.

### Conclusion

The analytical results make sense intuitively: if consumers do not consider that meat and cereals are easily substituable, they would want to consume a significant share of meat although its price is higher. On the contrary, if consumers are willing to substitute the two goods, then they would tend to consume the least expensive, i.e. cereals.

## Trade case
We assume no trade costs and impose the full immobility of lands (countries can not trade lands). The world is divided in two countries (or regions) R (rich) and P (poor) with same technologies, population normalized to 1, and $y_R \geq y_P$.

### Demand
Because agents have the same preferences in both countries, the objective function remains the same, only the budget constraint changes:
$$q_{c,w} + q_{m,w} p_w = y_R + y_P$$
Lands are fully immobile so income in each country remains $y_n = p_{L,n} \overline{L}_n$. 

### Production
While lands are fully immobile, cereals can be freely traded (because of no trade costs). Production functions are the same as under autarky, therefore the prices derived from the perfect competition assumption are:
$$ \forall n \in \left\lbrace R, P \right\rbrace, \hspace{3mm} p_{L,n} = \theta_c(d_n) $$
$$ p_w^\ast = \frac{1}{a} \left[ \eta_{m,R} \theta_c(d_R) + (1-\eta_{m,R}) \theta_c(d_P) \right] + \frac{1}{b} $$
Where $\eta_{m,R}= Q_{m,R}/Q_{m,w}$ is the share of country $R$ in total production of meat. In trade also, the optimal price of meat do not depend relative quantities.

### Equilibrium
Under trade, the equilibrium must satisfy the following constraints, the constraints on the full employment of resources are similar as in autarky. An additional constraint commed from the necessity of imports to equal exports:
$$ q_{c, P, R} + K_{m, P, R} + q_{c, R, P} + K_{m, R, P} = 0 $$
$$ q_{m, P, R} + q_{m, R, P} = 0 $$

From the expression of income and land price, we condlude that $y_R \geq y_P$ implies either that $\overline{L}_R \geq \overline{L}_P$ or $\theta_c(d_R) \geq \theta_c(d_P)$, or both.

In trade, the production possibility frontier requires the aggregate world crop productivity to be maximal, i.e.: $\theta_{c,w}= \theta_c(d_R)$ if $L_{c,w} < \overline{L}_R$, and $\theta_{c,w}=(\overline{L}_R/L_{c,w}) \theta_c(d_R) + ( 1 - \overline{L}_R/L_{c,w}) \theta_c(d_P) $ if $\overline{L}_R < L_{c,w} < \overline{L}_w$. Note that if both countries have same distance to the optimal temperature, hence same crop productivity, than the aggregate productivity is simply $\theta_{c,w}=\theta_c(d_R)=\theta_c(d_P)$.

This implies that the equation which led to the equilibrium quantities can not be solved by hand now, due to the aggregate productivity depending on $L_{c,w}$:
$$ a (\overline{L}_w - L_{c,w}) =  b (\theta_{c,w}(L_{c,w}) L_{c,w} - q_{c,w})$$
Therefore, we need numerical methods to derive the equilibrium quantities and the imports and exports quantities.

Nevertheless, with the expression of prices under autarky and trade, we can predict that:
$$ \forall n \in \left\lbrace R, P \right\rbrace, \hspace{3mm} p_n^\ast \geq p_w^\ast \Leftrightarrow \theta_{c} (d_n) \geq \theta_c(d_{n^\ast})$$
where $ n^\ast $ denotes the foreign country.
Unsurprisingly , one can also show that:
$$ \forall n \in \left\lbrace R, P \right\rbrace, \hspace{3mm} \frac{q_{m,n}}{q_{c,n}} \geq \frac{q_{m,w}}{q_{c,w}} \Leftrightarrow \theta_c(d_n) \leq \theta_c(d_{n^\ast})$$.

## Food security constraint
We check the satisfaction of subistence consumption ex-post. Denote $\underline{Q}$ the minimal quantity of calories (or any nutritional indicator) to eat for survival, and $M= \begin{pmatrix}
\mu_m \\
\mu_c
\end{pmatrix} \in [0;1]^2 $ the conversion matrix of kgs into calories (or any nutritional indicator). 

Subsistence consumption is satisfied if:
$$ (q_{m, n}^\ast \hspace{2mm} q_{c, n}^\ast) . M \geq \underline{Q} $$

Because we have the analytical expressions of the optimal quantities in autarky, we can derive the conditions for satsfying the food security conditions. Because $p^\ast \geq 1$, the condition is that the CES $\sigma$ is larger than the lower bound:
$$\sigma^\ast = \frac{1}{\ln(p^\ast)} \left[ \ln \left( \mu_m - \frac{\underline{Q} p^\ast}{\theta_c(d_n) \overline{L}_n} \right)  - \ln \left(\frac{\underline{Q}}{\theta_c(d_n) \overline{L}_n} - \mu_c \right) \right]$$

Which is defined if $\mu_m > \mu_c$ and relevant ($\sigma^\ast \in ]0;1[$) only if:
$$ (\mu_m + \mu_c) \frac{\theta_{c,n} \overline{L}_n}{\underline{Q}} - 1 > p^\ast > \exp \left( \ln \left( \mu_m - \frac{\underline{Q} p^\ast}{\theta_{c,n} \overline{L}_n} \right)  - \ln \left(\frac{\underline{Q}}{\theta_{c,n} \overline{L}_n} - \mu_c \right) \right) $$

Once again, for the trade case, we rely on numerical methods to find the solution.

# Computational approach

## Autarky
Parameters $a$, $b$, $\gamma_1$, $\gamma_2$ and $T^\ast$ are calibrated roughly with acutal data when available, generally using wheat as the representative cereal, and beef as the representative meat production.

The autarky module returns both analytical and numerical values. The analytical values use the expressions of otpimal quantities and prices derived in the model. The numerical values are obtained from the utility maximization subject to the budget constraint using the JuMP package. The results from this maximization depend on the optimal price for which we have the functional form. 

Because this module will be used by other modules, I created another module to plot the graphs so that ```include("autarkyeq.jl")``` does not take too much time.

## Trade
As explained, the equilibrium under trade is more complicated due to the expression of the aggregate crop productivity as a function of total lands required for growing crops. The same calibration is used in this module suited for worlds with two countries, or worlds with more countries as long as they all have different land endowment and thetas.

The expression of $\theta_{c,w}$ is a function taking as arguments $L_{c,w}$ and the matrix of countries characteristics sorted by decreasing thetas. When there are only 2 countries, there is no need for an optimization method with constraints: I use the package Optim to maximize $\theta_{c,w}$ with respect to $L_{c,R}$. When there are more than 2 countries, I use the JuMP package to maximize $\theta_{c,w}$. This function returns both the maximizers and the maximums, which allows to compute each country's share into total land allocated to crops, and total land allocated to meat. 

Note that the package Optim do not handle well the existence of several minimizer as it happens when thetas are the same across countries. Normally in that case, any allocation is optimal, but I manually force countries to allocate equally their shares of total land to crop production. I thus need to distinguish 2 cases : when the country with the smallest land endowment can welcome half the total required lands for crops, and when its land endwoment is not enough. This would make things a lot more complicated when there are more than 2 countries, so I just leave the JuMP package allocate land randomly in that case.

This allows to find the optimal relative price under trade, and therefore to run the utility maximization of the consumer. Due to the complicated expression of theta, utility maximization is now obtained with the package NLopt, where we define the objective function, the constraint and their gradients. Using JuMP would have make things unreadable.

The shares of countries into total land allocated to meat is equivalent to their share into total meat production because all countries have the same technology $a$. Thus, we can compute the optimal price as a function of $q_{c,w}$. I use the function ```fzeros()``` of the Roots package to find $L_{c,w}$ as a function of $q_{c,w}$. 

The module ```tradeeq``` computes the world equilibrium in trade, while the module ```tradeeqdf``` builds dataframes of both the results at the world level, and results at the country level such as the country's share in cereal production, land allocated to crops, its imports or exports of each product etc.

## Food security constraint

The module ```subsistconst.jl``` consists in a function that returns 1 if the constraint is satisfied, and 0 if not so that we can plot the results.

The parameters $\mu_m$, $\mu_c$, $\underline{Q}$ are defined at the individual level for one year using proteins as the nutritional unit and acutal data for the intake of proteins in wheat and beef, and the protein consumption recommandations for an average person.

Below, I plot results in a world with 2 countries with same CES = .8, and I make land endowment and temperature vary one after the other. For each world I plot one table with results at the country scale, and one table with global results.
